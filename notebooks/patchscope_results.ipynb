```python
import os
import sys
import pandas as pd

sys.path.append("../")
from proto import patchscope_pb2
```

```python
RESULT_FILENAMES = [
    "llama_3b__all_layers.binpb",
    "llama_8b__all_layers.binpb",
    "llama_3b_instruct__all_layers.binpb",
    "llama_8b_instruct__all_layers.binpb",
]

columns = ["result_set", "dataset", "accuracy"]
results_data = []
for filename in RESULT_FILENAMES:
    evaluation_results = patchscope_pb2.EvaluationResults()
    with open(os.path.join("results/", filename), "rb") as f:
        evaluation_results.ParseFromString(f.read())
    for result in evaluation_results.results:
        results_data.append([result.result_set_name, result.config.dataset, result.accuracy])

results_df = pd.DataFrame(results_data, columns=columns)
results_df
```
